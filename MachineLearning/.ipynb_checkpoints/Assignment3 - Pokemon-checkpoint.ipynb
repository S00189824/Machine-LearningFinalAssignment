# Reading in the data sets

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
#this will be my training set
df = pd.read_csv('pokemoninfo/PokemonDataSet.csv')
df2 = pd.read_csv('pokemoninfo/PokemonGen8.csv')
```

```python
print(df)
```

```python
#checking for any null values to see if data set is good
df.isnull().sum()
```

```python
#Dropped the columns according to the document
dset1 = df.drop(['Total','Name','ID','Type_1','Type_2','Generation','Legendary'], axis = 1)
```

```python
dset2 = df2.drop(["No.","Name","Ability1","Ability2","Hidden_Ability","Type1","Type2","Weight_kg","Height_m","Weight_lbs","Height_ft","Cap_Rate","Egg_Steps","Classification","Legendary"],axis=1)
```

```python
dset1YT_Legendary = df[['Legendary']].astype(int)# y train
```

```python
dset2_Legendary = df2[['Legendary']].astype(int)# y test
```

```python
dset1.head()
```

```python
X_train, X_val, Y_train,  Y_val = train_test_split(dset1, dset1YT_Legendary, test_size=0.25, random_state=1)
                                                   
data_to_be_set1 = X_train
scaler = MinMaxScaler()
scaler.fit(data_to_be_set1)
d_set1Normaized = scaler.transform(data_to_be_set1)# d_set1Normaized x_train
data_to_be_set2=dset2
scaler.fit(data_to_be_set2)
Set2_Normalize=scaler.transform(data_to_be_set2)# Set2_Normalize Test
```

# Sequencial Model 1

```python
sqmodel = Sequential()
```

```python
sqmodel.add(Dense(32, activation='relu', input_dim = 6))# input dimensions
sqmodel.add(Dense(1, activation='sigmoid'))# input dimensions
```

```python
sqmodel.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3), #learning rate
              loss=tf.keras.losses.BinaryCrossentropy(), # measuring thee loss
              metrics=[tf.keras.metrics.BinaryAccuracy(), # measuring the accuracy true/false
                       tf.keras.metrics.FalseNegatives()]) # measuring how many false negatives
```

```python
md1 = sqmodel.fit(d_set1Normaized, Y_train, 
            epochs = 500, callbacks = tf.keras.callbacks.EarlyStopping(monitor = 'binary_accuracy', 
            patience = 100),validation_split=0.0,validation_data = (X_val,Y_val))
```

```python
loss = sqmodel.evaluate(d_set1Normaized, Y_train)
```

```python
testresults = sqmodel.predict(Set2_Normalize)
print (testresults)
```

```python
# 
ds = np.array(testresults)

ds1 = ds.reshape(400,1)

Gen8Pokemon = np.append(ds1, df2[['Name']],axis=1)
print_stat_greater_than = Gen8Pokemon[Gen8Pokemon[:,0]>0.95 ]# Print pokemon greater than 95% with  names
print(print_stat_greater_than)
```

# Sequencial Model 2

```python
#Defining the model

model2 = Sequential()
model2.add(Dense(32, activation='tanh', input_dim = 6))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# early Stopping
es = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', mode='min', verbose=1, patience=100)

# fitting model
md2 = model2.fit(d_set1Normaized, Y_train, validation_data = (X_val,Y_val), epochs=500, verbose=1, callbacks=[es])

#evaluating model
_, train_acc = model2.evaluate(d_set1Normaized, Y_train, verbose=0)
_, test_acc = model2.evaluate(Set2_Normalize, dset2_Legendary, verbose=0)
```
